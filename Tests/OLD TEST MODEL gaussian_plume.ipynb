```python
import numpy as np
import matplotlib.pyplot as plt

# Constants and parameters (добавьте ваши значения здесь)
Q = 10    # source emission rate (kg/s)
U = 3     # wind speed (m/s)
K = 5     # eddy diffusion coefficient (m^2/s)
min_concentration = 1e-25  # Минимальное значение концентрации
rmax = 4000  # size of domain in r (m)
ymax = 2000  # size of domain in y (m)
N = 2000     # number of plotting points

def gauss_func(U, xx, yy):
    xx = np.where(xx == 0, 0.005, xx)  # Избежать деления на ноль
    return Q / (2 * np.pi * K * xx) * np.exp(-U * yy**2 / (4 * K * xx))
```

```python
# Уровни для графиков
lev = list(np.geomspace(min_concentration, 1, 20))
# Создание подграфиков
fig, axes = plt.subplots(1, 2, figsize=(15, 5), sharey=True)  # 1 строка, 3 столбца




# Первый график
Q = 10    # source emission rate (kg/s)
U = 3     # wind speed (m/s)
K = 5     # eddy diffusion coefficient (m^2/s)

rr, yy = np.meshgrid(np.linspace(0.005, rmax, N), np.linspace(-ymax, ymax, 2 * N))
cs = axes[0].contourf(rr, yy, gauss_func(U, rr, yy), lev, colors=['#808080', '#A0A0A0', '#C0C0C0'])
axes[0].set_title('Q = 10')




# Второй график
Q = 100    # source emission rate (kg/s)
U = 3     # wind speed (m/s)
K = 5     # eddy diffusion coefficient (m^2/s)

cs = axes[1].contourf(rr, yy, gauss_func(U, rr, yy), lev, colors=['#808080', '#A0A0A0', '#C0C0C0'])
axes[1].set_title('Q = 100')





# Общие настройки
plt.subplots_adjust(right=0.85)  # Уменьшение отступов справа для цветовой шкалы

# Добавление цветовой шкалы
cbar = plt.colorbar(cs, ax=axes, orientation='vertical')
cbar.ax.set_yscale('log')  # Логарифмическая шкала
cbar.set_label('Концентрация (логарифм)')

# Сохранение и отображение
plt.savefig('multiple_plots.jpg', dpi=300, bbox_inches='tight', facecolor="white")
plt.show()
```
